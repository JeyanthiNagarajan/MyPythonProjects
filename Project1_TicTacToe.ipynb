```python
## Get the player's name
print "WELCOME TO TIC-TAC_TOE!!!"
player1 = raw_input("Type Player 1 Name? ")
player2 = raw_input("Type Player 2 Name? ")
#Create a empty Matrix to play Tic Tac Toe
TicTacToe = [[' ' for row in range(3)] for col in range(3)] 
#TicTacToe = [[1,2,3], [4,5,6], [7,8,9]] 

#Function to Display the board
from IPython.display import clear_output
def DisplayTicTacToe_Board():
    clear_output()
    print "\n * Check Out the current TicTacToe Board Status *"
    print('   |   |')
    print(' ' + TicTacToe[0][0] + ' | ' + TicTacToe[0][1] + ' | ' + TicTacToe[0][2])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + TicTacToe[1][0] + ' | ' + TicTacToe[1][1] + ' | ' + TicTacToe[1][2])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + TicTacToe[2][0] + ' | ' + TicTacToe[2][1] + ' | ' + TicTacToe[2][2])
    print('   |   |')

    
#Function to assign the user input to board
def AssignTicTacToe_Board(player, row, col) :
    
    #Validate if the index is within the limit
    if row >= 0 and row <=2 and col >=0 and col <=2:
        
        #Check if the given index is already occupied :
        if TicTacToe[row][col] != 'X' and TicTacToe[row][col] !='O' :
            
            if player == player1:
                TicTacToe[row][col] = 'X'
                return True
            elif player == player2:
                TicTacToe[row][col] = 'O'
                return True
            else:
                print "ERROR : Only %s or %s are allowed to play \n " %(player1, player2)
                return False
        else :
            
            print "ERROR : This Position is already Occupied. Try Different Position !!!"
            return False
    
    else:
        print "ERROR : Index Out Of Range, TRY AGAIN!!! \n"
        return False


#Function to check if any player succeeded
def ValidateTicTacToe_Board(player):
    
    Marker = ' '
    if player == player1:
        Marker = 'X'
    else :
        Marker = 'O'
    
    return((TicTacToe[0][0] == TicTacToe[0][1] == TicTacToe[0][2] == Marker) or
          (TicTacToe[1][0] == TicTacToe[1][1] == TicTacToe[1][2] == Marker) or
          (TicTacToe[2][0] == TicTacToe[2][1] == TicTacToe[2][2] == Marker) or
          (TicTacToe[0][0] == TicTacToe[1][0] == TicTacToe[2][0] == Marker) or
          (TicTacToe[0][1] == TicTacToe[1][1] == TicTacToe[2][1] == Marker) or
          (TicTacToe[0][2] == TicTacToe[1][2] == TicTacToe[2][2] == Marker) or
          (TicTacToe[0][0] == TicTacToe[1][1] == TicTacToe[2][2] == Marker) or
          (TicTacToe[2][0] == TicTacToe[1][1] == TicTacToe[0][2] == Marker))
        
    
def ReplayTicTacToe():
    
    #replay = input("Would you like to replay the game ? Press 1 for Yes, 0 for No ")
    return raw_input('Do you want to play again? Enter Yes or No: ').lower().startswith('y')
    #return replay
        
def MainTicTacToe():
    
    #Play the game until 9 moves or any player wins
    GameOn = True
    NoOfMoves = 1
    global TicTacToe

    while GameOn :
        
        #Play until 9 moves
        if NoOfMoves <= 9 :
            
            #If the current move is odd no, that means it's Player1's turn. Otherwise its Player2's turn.
            if NoOfMoves % 2 != 0 :
                print "\n ***** IT'S %s'S TURN ***** \n" %(player1)
                player = player1
            else :
                print "\n ***** IT'S %s'S TURN ***** \n" %(player2)
                player = player2
                
            row = input("Enter the row you want to place the Mark :") 
            col = input("Enter the column you want to place the Mark :") 
                
            if AssignTicTacToe_Board(player, row, col) == True :
                NoOfMoves += 1
                DisplayTicTacToe_Board()
        
                if ValidateTicTacToe_Board(player) == True :
                    print '\n ***** CONGRATULATIONS!!! %s WON THE GAME ***** \n' %(player)
                    
                    #Check if the players wants to replay the game
                    if ReplayTicTacToe() :
                        #Reset the No Of moves and the board
                        NoOfMoves = 1
                        TicTacToe = [[' ' for row in range(3)] for col in range(3)] 
                    
                    else:
                        GameOn = False
                        print "Bye Bye... Have a Good Day !!!"
                else :
                    pass
            else :
                pass
                
        #If 9 moves elapsed and no one won, then game is a draw.
        else :
            print "IT'S A DRAW"
            #Check if the players wants to replay the game
            if ReplayTicTacToe() :
                #Reset the No Of moves and the board
                NoOfMoves = 1
                TicTacToe = [[' ' for row in range(3)] for col in range(3)] 
                    
            else:
                GameOn = False
                print "Bye Bye... Have a Good Day !!!"
            
#Call the Main function
MainTicTacToe()


            
```
